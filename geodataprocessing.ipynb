```python
import geopandas as gpd
import os
import fiona
import pandas as pd
import numpy as np
```

```python
dir_name= './bangladesh_latest_osm/'
extension = ".shp"

shapefiles = []

for item in os.listdir(dir_name): # loop through items in dir
    if item.endswith(extension): # check for ".shp" extension
        file_name = os.path.join(dir_name, item)
        shapefiles.append(gpd.read_file(file_name)) 
```

```python
for shapefile in shapefiles:
    print(shapefile.bounds)
```
